```python
import numpy as np
import matplotlib.pyplot as plt

# Homework 2 a)

a = -101/100
n = np.arange(1., 500, 1)
x = a**n / n 
b = np.cumsum(x)

plt.scatter(n, b)

plt.xlabel("n")
plt.ylabel("$(-101/100)^n/n$")
```
